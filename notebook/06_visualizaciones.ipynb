```python
import matplotlib.pyplot as plt
```

```python
model_names = ["MaskPredict", "Fertility", "LatentNAT"]
bleu_scores = [22.4, 24.1, 25.3]
latencies = [0.023, 0.035, 0.041]

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.bar(model_names, bleu_scores, color='skyblue')
plt.title("BLEU Score por modelo")
plt.ylabel("BLEU")

plt.subplot(1, 2, 2)
plt.bar(model_names, latencies, color='salmon')
plt.title("Latencia promedio")
plt.ylabel("Segundos")

plt.tight_layout()
plt.show()
```
